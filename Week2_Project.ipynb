Note the target column name and the ID column name.

```python
possible_id_names = [c for c in train.columns if c.lower() in ('id','index','idx')]
id_col = possible_id_names[0] if possible_id_names else (train.columns[0] if train.iloc[:,0].is_unique else None)


train_num = train.select_dtypes(include=[np.number]).columns.tolist()
test_num = test.select_dtypes(include=[np.number]).columns.tolist()
candidate_targets = [c for c in train_num if c not in test_num]
if len(candidate_targets) == 1:
   target_col = candidate_targets[0]
elif len(candidate_targets) > 1:
   target_col = candidate_targets[-1]

else:
   target_col = train_num[-1]


print('Detected id_col =', id_col)
print('Detected target_col =', target_col)
```

Feature list

```python
numeric_features = [c for c in train.select_dtypes(include=[np.number]).columns if c not in ([id_col, target_col] if id_col is not None else [target_col])]
categorical_features = train.select_dtypes(include=['object','category']).columns.tolist()
```

Drop Constant cols

```python
const_cols = [c for c in train.columns if train[c].nunique() <= 1]
if const_cols:
   train.drop(columns=const_cols, inplace=True)
   test.drop(columns=const_cols, inplace=True)
   numeric_features = [c for c in numeric_features if c not in const_cols]
   categorical_features = [c for c in categorical_features if c not in const_cols]


print('\nNumeric features count:', len(numeric_features))
print('Categorical features count:', len(categorical_features))
```

frequency encoding for large-cardinality

```python
ohe_threshold = 20
ohe_features = [c for c in categorical_features if train[c].nunique() <= ohe_threshold]
large_cat_features = [c for c in categorical_features if c not in ohe_features]


print('\nOHE features (<=20 unique):', len(ohe_features))
print('Large categorical features (freq-encoded):', len(large_cat_features))
```

Frequency encode large categorical features

```python
def freq_encode(df, cols):
  df = df.copy()
  for c in cols:
      freq = df[c].value_counts(normalize=True)
      df[c + '_freq_enc'] = df[c].map(freq).fillna(0.0)
  return df


if large_cat_features:
   train = freq_encode(train, large_cat_features)
   test = freq_encode(test, large_cat_features)
   for c in large_cat_features:
       numeric_features.append(c + '_freq_enc')
```

```python
# Update categorical_features to only those we will perform OHE
categorical_features = ohe_features
```

Preprocessing pipelines

```python
num_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')), ('var', VarianceThreshold()), ('scale', StandardScaler())])
cat_pipeline = Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])


preprocessor = ColumnTransformer(transformers=[('num', num_pipeline, numeric_features), ('cat', cat_pipeline, categorical_features)], remainder='drop', sparse_threshold=0)
```

```python
features = [c for c in train.columns if c not in ([id_col, target_col] if id_col is not None else [target_col])]
X = train[features].replace([np.inf, -np.inf], np.nan)
y = train[target_col]
```

```python
# Baseline
baseline_mae = mean_absolute_error(y, np.full_like(y, y.mean()))
print('\nBaseline MAE (train mean):', baseline_mae)
```

Regression

```python
from sklearn.model_selection import cross_val_score
```

```python
models = {
'Linear': Pipeline([('pre', preprocessor), ('est', LinearRegression())]),
'Ridge': Pipeline([('pre', preprocessor), ('est', Ridge())]),
'Lasso': Pipeline([('pre', preprocessor), ('est', Lasso(max_iter=10000))]),
'ElasticNet': Pipeline([('pre', preprocessor), ('est', ElasticNet(max_iter=10000))])
}


def cv_mae(pipe, X, y, cv=5):
    scores = cross_val_score(pipe, X, y, cv=cv, scoring='neg_mean_absolute_error', n_jobs=-1)
    return -scores.mean(), -scores.std()


for name, pipe in models.items():
    mean_mae, std_mae = cv_mae(pipe, X, y, cv=5)
    print(f"{name}: CV MAE = {mean_mae:.4f} ± {std_mae:.4f}")
```

Grid search tuning for Ridge, Lasso, ElasticNet

```python
param_grids = {
'Ridge': {'est__alpha': [0.01, 0.1, 1, 10, 50, 100]},
'Lasso': {'est__alpha': [0.0005, 0.001, 0.01, 0.1, 1, 10]},
'ElasticNet': {'est__alpha': [0.0005, 0.001, 0.01, 0.1, 1], 'est__l1_ratio': [0.2, 0.5, 0.8]}
}


best_models = {}
for name in ['Ridge','Lasso','ElasticNet']:
    print('\nGrid searching', name)
    grid = GridSearchCV(models[name], param_grids[name], cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)
    grid.fit(X, y)
    best_models[name] = grid.best_estimator_
    print(name, 'best CV MAE =', -grid.best_score_, 'params =', grid.best_params_)
```

Compare Final Cross Validation Scores

```python
all_results = []
for name, pipe in [('Linear', models['Linear']), ('Ridge', best_models['Ridge']), ('Lasso', best_models['Lasso']), ('ElasticNet', best_models['ElasticNet'])]:
    mean_mae, std_mae = cv_mae(pipe, X, y, cv=5)
    all_results.append({'model': name, 'mae_mean': mean_mae, 'mae_std': std_mae})


results_df = pd.DataFrame(all_results).sort_values('mae_mean')
print('\nCV comparison:')
print(results_df)
```

Choose best, retrain on full train, predict test

```python
best_model_name = results_df.iloc[0]['model']
print('\nChosen model:', best_model_name)
if best_model_name == 'Linear':
   final_pipe = models['Linear']
elif best_model_name == 'Ridge':
    final_pipe = best_models['Ridge']
elif best_model_name == 'Lasso':
    final_pipe = best_models['Lasso']
else:
    final_pipe = best_models['ElasticNet']


final_pipe.fit(X, y)
```

Prepare test X

```python
test_X = test[features].replace([np.inf, -np.inf], np.nan)
preds = final_pipe.predict(test_X)
```

Submission

```python
if id_col is not None and id_col in test.columns:
   ids = test[id_col]
else:
   ids = np.arange(1, len(test) + 1)
submission = pd.DataFrame({id_col if id_col is not None else 'ID': ids, 'Predicted': preds})
submission.to_csv(OUT_SUBMISSION, index=False)
print('Saved submission to', OUT_SUBMISSION)
```
